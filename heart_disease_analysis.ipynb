```python
# Patient Dataset
import numpy as np #linear algebra
import pandas as pd #data processing, csv file
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/heart.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
df.describe()
```

```python
#target distribution
df.target.value_counts()
```

```python
#show correaltion matrix
corr=df.corr()
plt.figure(figsize=(18,10))
sns.heatmap(corr,annot=True, )
plt.show()
```

```python
sns.countplot(x=df.target,hue='sex',data=df,palette=['#1CA53B','red'])
plt.xticks([0,1],labels=['Not Disease','Disease'])
plt.title("Target Distribution")
```

```python
plt.figure(figsize=(18, 10))
sns.countplot(x='age', hue='sex', data=df, palette=['#1CA53B', 'red'])
plt.legend(["Haven't Disease","Have Disease"])
plt.title("Heart Disease Frequency for ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#sns.set_style("white grid")
plt.figure(figsize=(18,10))
sns.distplot(df.age[df['target']==0],bins=30,color='red',label='Haven\'t Disease')
sns.distplot(df.age[df['target']==1],bins=30,color='green',label='Have Disease')
plt.legend()
plt.title("Heart Disease Frequency for Ages")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
